```python
# Hyper Parameters for packages
nlp = spacy.load("en_core_web_lg")

language = "en"
kw_extractor = yake.KeywordExtractor()
max_ngram_size = 4
deduplication_threshold = 0.3
numOfKeywords = 10
custom_kw_extractor = yake.KeywordExtractor(lan=language,
                                            n=max_ngram_size,
                                            dedupLim=deduplication_threshold,
                                            top=numOfKeywords,
                                            features=None)
```

```python
def extract_named_entities(text):
    # https://spacy.io/usage/linguistic-features#named-entities
    if isinstance(text, list):
        text = ' '.join(x for x in text)
        
    doc = nlp(text)
    ents = []
    for ent in doc.ents:
        ents.append([ent.text, ent.start_char, ent.end_char, ent.label_])

    return ents

def get_wiki(topic):
    # https://pypi.org/project/wikipedia/
    search = wiki.search(topic, results= 50, suggestion=True)
    print("\n--WIKIPEDIA PAGE {}--\n".format(search[0]))
    
    page = wiki.WikipediaPage(search[0])
    categories = page.categories
    sections = page.sections
    summary = page.summary
    
    text = wiki.summary(search[0])
    suggest = wiki.suggest(topic)
    
    

    return (text, search, suggest, categories, sections, summary)


def yake_keywords(text):
    # https://towardsdatascience.com/keyword-extraction-process-in-python-with-natural-language-processing-nlp-d769a9069d5c
    keywords = custom_kw_extractor.extract_keywords(text)

    return keywords
```

```python
graph = Network(height='800px',width='1200px',directed=False)

text, search, suggest, categories, sections, summary = get_wiki('mcdonalds')

yake_kw = yake_keywords(text)
ents = extract_named_entities(text)


print('--YAKE KEYWORDS--')
print(yake_kw, '\n')
print('\n--SPACY NAMED ENTITIES--')
print(ents, '\n')
print('\n--WIKIPEDIA SEARCH--')
print(search, '\n')
print('\n--WIKIPEEDIA SUGGEST--')
print(suggest, '\n')
print('\n--WIKIPEEDIA TEXT--')
print(text, '\n')

print('\n--WIKIPEEDIA CATEGORIES--')
print(categories, '\n')
print('\n--WIKIPEEDIA SECTIONS--')
print(sections, '\n')
print('\n--WIKIPEEDIA SUMMARY--')
print(summary, '\n')
```

```python
def find_substring_in_array_of_strings(arrayS1, arrayS2):
    a1 = [x.split(' ') for x in arrayS1]
    a2 = [x.split(' ') for x in arrayS2]
    
    w1 = []
    w2 = []
    
    matching = []
    
    for topic in a1:
        for word in topic:
            w1.append(word.lower())
                
    for topic in a2:
        for word in topic:
            w2.append(word.lower())
    
    for l_word in w1:
        if l_word in w2:
            matching.append(l_word)
            print('--', l_word)
    
    no_stop = [word for word in matching if not word in stopwords.words('english')]

    return set(no_stop)
            
    

```

```python
def compare_two_topics(t1, t2):
    topic1_set =  wiki.search(t1, results=25)
    topic2_set = wiki.search(t2, results=25)
    
    return find_substring_in_array_of_strings(topic1_set, topic2_set)
                
```

```python
print(compare_two_topics('Die Hard', 'Pulp Fiction'))
```
